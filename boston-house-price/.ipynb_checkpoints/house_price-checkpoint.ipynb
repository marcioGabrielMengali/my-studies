## DATA PREPROCESSING

```python
chas_column = dataset.pop('CHAS')
```

```python
dataset.columns
```

```python
dataset.insert(0,'CHAS',chas_column)
```

```python
dataset.columns
```

```python
X = dataset.drop('MEDV',axis=1).values
```

```python
print(X.shape)
```

```python
y = dataset['MEDV'].values
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
print(X_train)
```

```python
#Standar Scaler X
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train[:,1:] = sc_x.fit_transform(X_train[:,1:])
X_test[:,1:] = sc_x.transform(X_test[:,1:])
print(X_train)
print(X_test)
```

```python
#Standar Scaler Y
y_train = y_train.reshape((-1,1))
y_test = y_test.reshape((-1,1))
print(y_train)
```

```python
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
print(y_train)
print(y_test)
```

# TRAINING MODELS

## LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
regression = LinearRegression()
regression.fit(X_train, y_train)
print(f'Regression Score {regression.score(X_test,y_test)}')
kf = KFold(n_splits=6,shuffle=True,random_state=42)
cv_scores = cross_val_score(regression, X, y,cv=kf)
print(cv_scores)
```

## SUPORT VECTOR REGRESSION

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR
regression = SVR()
params = {
    'C': np.linspace(0.1,1,50),
    'kernel': ['rbf', 'poly','sigmoid','linear'],
    'degree': [1,2,3,4,5,6],
    'gamma': np.linspace(0.0001, 1.0, 50)
}
regression_cv = RandomizedSearchCV(
    regression,
    params,
    cv=kf,
    n_iter = 10
)
regression_cv.fit(X_train,y_train)
```

```python
print(regression_cv.best_params_)
```

```python
print(regression_cv.best_score_)
```

## DECISION TREE REGRESSION

```python
from sklearn.tree import DecisionTreeRegressor
regression = DecisionTreeRegressor()
params = {
    'splitter': ['best','random'],
    "criterion": ["mse", "mae"],
    "min_samples_split": [10, 20, 40],
    "max_depth": [2, 6, 8],
    "min_samples_leaf": [20, 40, 100],
    "max_leaf_nodes": [5, 20, 100],
              }
regression_cv = RandomizedSearchCV(
    regression,
    params,
    cv=kf,
    n_iter = 10
)
regression_cv.fit(X_train, y_train)
```

```python
print(regression_cv.best_params_)
```

```python
print(regression_cv.best_score_)
```

## RANDOM FOREST REGRESSION

```python
from sklearn.ensemble import RandomForestRegressor
regression = RandomForestRegressor()
params = {
    'n_estimators': [int(x) for x in np.linspace(200,2000,10)],
    "max_features": ['auto','sqrt'],
    "max_depth": np.linspace(10,110,10),
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4],
    'bootstrap': 'bootstrap'
              }
regression_cv = RandomizedSearchCV(
    regression,
    params,
    cv=kf,
    n_iter = 10
)
regression_cv.fit(X_train, y_train)
```

```python
print(regression_cv.best_params_)
```

```python
print(regression_cv.best_score_)
```

## RIDGE

```python
from sklearn.linear_model import Ridge
regression = Ridge()
params = {
    'alpha': np.linspace(0.1,100,50)
    }
regression_cv = RandomizedSearchCV(
    regression,
    params,
    cv=kf,
    n_iter = 10
)
regression_cv.fit(X_train, y_train)
```

```python
print(regression_cv.best_params_)
```

```python
print(regression_cv.best_score_)
```

## LASSO

```python
from sklearn.linear_model import Lasso
regression = Lasso()
params = {
    'alpha': np.linspace(0.1,100,50)
    }
regression_cv = RandomizedSearchCV(
    regression,
    params,
    cv=kf,
    n_iter = 10
)
regression_cv.fit(X_train, y_train)
```

```python
print(regression_cv.best_params_)
```

```python
print(regression_cv.best_score_)
```

```python
names = dataset.drop('MEDV',axis=1).columns
lasso = Lasso(alpha=0.1)
lasso_coef = lasso.fit(X,y).coef_
plt.bar(names,lasso_coef)
plt.xticks(rotation=45)
plt.show()
```
